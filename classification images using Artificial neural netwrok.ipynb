```python
def get_model():
    model=keras.Sequential([
               keras.layers.Flatten(input_shape=(32,32,3)),# input shape is 32*32*3=3072
               keras.layers.Dense(3000,activation='relu'),#Hidden layers
               keras.layers.Dense(3000,activation='relu'),#Hidden layers,more hidden  layers tends to work better
               keras.layers.Dense(10,activation='sigmoid')#output layers
    ])
    
    model.compile(optimizer='SGD',# here we use Stichastic gradient descent because data is large
             loss='categorical_crossentropy',
             metrics=['accuracy'])
    
    return model    
```

```python
%%timeit -n1 -r1  # for measuring the time with cpu
with tf.device('/CPU:0'):
    cpu_model=get_model()
    cpu_model.fit(x_train,y_train_categorical,epochs=1)
```
